```python
numRows = int(input())
result = []
if numRows == 0:
    print(result)

first_row = [1]
result.append(first_row)

for i in range(1, numRows):
    current_row = [1]
    prev_row = result[i - 1]

    for j in range(1, i):
        current_row.append(prev_row[j - 1] + prev_row[j])

    current_row.append(1)
    result.append(current_row)

print(result)
```

```python
rowIndex = int(input())

result = []

first_row = [1]
result.append(first_row)

if rowIndex == 0:
    print(result)

for i in range(1, rowIndex + 1):
    current_row = [1]
    prev_row = result[i - 1]

    for j in range(1, i):
        current_row.append(prev_row[j - 1] + prev_row[j])

    current_row.append(1)
    result.append(current_row)

print(current_row)
```
