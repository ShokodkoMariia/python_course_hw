```python
words = ["Hello","Alaska","Dad","Peace"]
s1 = 'qwertyuiop'
s11 = s1.upper() + s1
s2 = 'asdfghjkl'
s22 = s2.upper() + s2
s3 = 'zxcvbnm'
s33 = s3.upper() + s3
result = []
for word in words:
    if set(word) - set(s11) == set():
        result.append(word)
    elif set(word) - set(s22) == set():
        result.append(word)
    elif set(word) - set(s33) == set():
        result.append(word)
print(result)
```
